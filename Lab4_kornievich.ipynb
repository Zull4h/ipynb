```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
titanic = sns.load_dataset("titanic")
titanic.info()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 4, figsize=(16, 10), sharex=False, sharey=False)

sns.countplot(x='pclass', data=titanic, ax=axes[0,0])
sns.countplot(x='survived', data=titanic, ax=axes[0,1])
sns.countplot(x='sex', data=titanic, ax=axes[0,2])
sns.countplot(x='sibsp', data=titanic, ax=axes[0,3])
sns.countplot(x='embarked', data=titanic, ax=axes[1,0])
sns.countplot(x='parch', data=titanic, ax=axes[1,1])
sns.distplot(titanic['fare'], ax= axes[1,2])
sns.distplot(titanic['age'].dropna(), ax = axes [1,3])
```

```python
titanic_corr = titanic.corr()
```

```python
fig2, axes2 = plt.subplots(2, 3, figsize=(16, 10), sharex=False, sharey=False)

sns.stripplot(x="sex", y="age", data=titanic , jitter=True ,ax=axes2[0,0])
sns.barplot(x="pclass", y="survived", data=titanic, ax=axes2[0,1])
sns.barplot(x="embarked", y="survived", data=titanic, ax = axes2 [0,2])
sns.barplot(x="pclass", y="age", data=titanic, ax = axes2[1,1])
sns.barplot(x='embarked', y='fare', data=titanic, ax = axes2[1,2])
sns.barplot(x='sex', y='survived', data=titanic, ax = axes2[1,0])
```
